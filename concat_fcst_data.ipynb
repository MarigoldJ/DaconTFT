```python
import os
import pandas as pd
```

```python
# 여러 예보 데이터들을 모아둘 list 만들기
list_ulsan = []
list_dangjin = []
```

```python
# 직접 가져와서 정리한 2~6월 예보 데이터 추가
dir_data = './new_dataset_v2'

df = pd.read_csv(os.path.join(dir_data, 'new_fcst_dangjin_6.csv'))
list_dangjin.append(df)

df = pd.read_csv(os.path.join(dir_data, 'new_fcst_ulsan_6.csv'))
list_ulsan.append(df)
```

```python
# api로 가져온 7월 예보
dir_api = './fcst_api_data'
list_api_data = os.listdir(dir_api)

for f in list_api_data:
    if 'dangjin' in f:
        print(f'{f} -> dangjin')
        df = pd.read_csv(os.path.join(dir_api, f), index_col=None)
        list_dangjin.append(df)
    elif 'ulsan' in f:
        print(f'{f} -> ulsan')
        df = pd.read_csv(os.path.join(dir_api, f), index_col=None)
        list_ulsan.append(df)

        
```

```python
df_ulsan = pd.DataFrame()
for ulsan in list_ulsan:
    df_ulsan = pd.concat([df_ulsan, ulsan])

df_dangjin = pd.DataFrame()
for dangjin in list_dangjin:
    df_dangjin = pd.concat([df_dangjin, dangjin])
```
