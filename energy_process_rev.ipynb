```python
import pandas as pd
```

```python
url_csv1 = 'https://raw.githubusercontent.com/LeeSH825/TemporalFusionTransformer/master/csv_data/%ED%95%9C%EA%B5%AD%EB%8F%99%EC%84%9C%EB%B0%9C%EC%A0%84%20%EC%8B%9C%EA%B0%84%EB%8C%80%EB%B3%84%20%ED%83%9C%EC%96%91%EA%B4%91%20%EB%B0%8F%20%ED%92%8D%EB%A0%A5%20%EB%B0%9C%EC%A0%84%EB%9F%89%20%ED%98%84%ED%99%A9(2018_2019)_utf.csv'
url_csv2 = 'https://raw.githubusercontent.com/LeeSH825/TemporalFusionTransformer/master/csv_data/%ED%95%9C%EA%B5%AD%EB%8F%99%EC%84%9C%EB%B0%9C%EC%A0%84%20%EC%8B%9C%EA%B0%84%EB%8C%80%EB%B3%84%20%ED%83%9C%EC%96%91%EA%B4%91%20%EB%B0%9C%EC%A0%84%EB%9F%89%20%ED%98%84%ED%99%A9(2015_2017)_utf.csv'

past_energy1 = pd.read_csv(url_csv1)
past_energy2 = pd.read_csv(url_csv2)
energy = pd.read_csv('./dataset/energy.csv')
```

```python
energy
```

# 1.2018년 데이터 정리

```python
past_energy1['태양광명'].value_counts()
# 울산태양광
# 당진자재창고태양광
# 당진수상태양광
# 당진태양광
```

```python
df_ulsan1 = past_energy1[past_energy1['태양광명'] == '울산태양광']
df_dj_warehouse1 = past_energy1[past_energy1['태양광명'] == '당진자재창고태양광']
df_dj_floating1 = past_energy1[past_energy1['태양광명'] == '당진수상태양광']
df_dj1 = past_energy1[past_energy1['태양광명'] == '당진태양광']

print(df_ulsan1.iloc[0, 1], df_ulsan1.iloc[-1, 1])
print(df_dj_warehouse1.iloc[0, 1], df_dj_warehouse1.iloc[-1, 1])
print(df_dj_floating1.iloc[0, 1], df_dj_floating1.iloc[-1, 1])
print(df_dj1.iloc[0, 1], df_dj1.iloc[-1, 1])
print()

# 우리가 필요한건 2018-01-01 ~ 2018-02-28 (59개 행만 필요)
df_ulsan1 = df_ulsan1.iloc[:59]
df_dj_warehouse1 = df_dj_warehouse1.iloc[:59]
df_dj_floating1 = df_dj_floating1.iloc[:59]
df_dj1 = df_dj1.iloc[:59]

print(df_ulsan1.iloc[0, 1], df_ulsan1.iloc[-1, 1])
print(df_dj_warehouse1.iloc[0, 1], df_dj_warehouse1.iloc[-1, 1])
print(df_dj_floating1.iloc[0, 1], df_dj_floating1.iloc[-1, 1])
print(df_dj1.iloc[0, 1], df_dj1.iloc[-1, 1])
```

```python
ulsan1 = pd.DataFrame(columns=['time', 'ulsan'])
dj_warehouse1 = pd.DataFrame(columns=['time', 'dangjin_warehouse'])
dj_floating1 = pd.DataFrame(columns=['time', 'dangjin_floating'])
dj1 = pd.DataFrame(columns=['time', 'dangjin'])

for i in range(len(df_ulsan1)):
    row = df_ulsan1.iloc[i].copy()
    temp = pd.DataFrame(row[2:])
    date = row.iloc[1]

    temp.reset_index(inplace=True)
    temp.columns = ['hour', 'ulsan']
    temp['date'] = date
    temp['time'] = temp['date'] + ' ' + temp['hour'] + ':00:00'
    temp = temp[['time', 'ulsan']]
    
    ulsan1 = pd.concat([ulsan1, temp])

for i in range(len(df_dj_warehouse1)):
    row = df_dj_warehouse1.iloc[i].copy()
    temp = pd.DataFrame(row[2:])
    date = row.iloc[1]

    temp.reset_index(inplace=True)
    temp.columns = ['hour', 'dangjin_warehouse']
    temp['date'] = date
    temp['time'] = temp['date'] + ' ' + temp['hour'] + ':00:00'
    temp = temp[['time', 'dangjin_warehouse']]
    
    dj_warehouse1 = pd.concat([dj_warehouse1, temp])

for i in range(len(df_dj_floating1)):
    row = df_dj_floating1.iloc[i].copy()
    temp = pd.DataFrame(row[2:])
    date = row.iloc[1]

    temp.reset_index(inplace=True)
    temp.columns = ['hour', 'dangjin_floating']
    temp['date'] = date
    temp['time'] = temp['date'] + ' ' + temp['hour'] + ':00:00'
    temp = temp[['time', 'dangjin_floating']]
    
    dj_floating1 = pd.concat([dj_floating1, temp])
    
for i in range(len(df_dj1)):
    row = df_dj1.iloc[i].copy()
    temp = pd.DataFrame(row[2:])
    date = row.iloc[1]

    temp.reset_index(inplace=True)
    temp.columns = ['hour', 'dangjin']
    temp['date'] = date
    temp['time'] = temp['date'] + ' ' + temp['hour'] + ':00:00'
    temp = temp[['time', 'dangjin']]
    
    dj1 = pd.concat([dj1, temp])
```

```python
energy_18 = pd.merge(ulsan1, dj_warehouse1, how='left', on='time')
energy_18 = pd.merge(energy_18, dj_floating1, how='left', on='time')
energy_18 = pd.merge(energy_18, dj1, how='left', on='time')
energy_18
```
